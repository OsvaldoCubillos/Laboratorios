## Ejercicio 1

Escribir un programa que pregunte al usuario o genere las ventas de los últimos 5 días, y muestre por pantalla una serie (Dataframe) con los datos de las ventas por producto (opcional): 
* Nro de unidades y precio unitario, indexada por los días, antes y después de aplicarles un descuento del 5%.
(Diversas opciones de solución: libre)

### Solución
Lo primero que se realizará será preparar al sistema para que compila la información, por ello, se cargarán las librerías de `pandas`-`nunpy` y `datatime`. Además se crea un `dataframe` vacío.
Finalmente, se crean variables para recopilar la información con los nombre:
* producto
* cantidad
* ventas

```python
# Importando librería de Pandas para trabajar con DataFrame
import pandas as pd
import numpy as np

# librería para el ingreso de la fecha
from datetime import datetime 

#creando DataFrame vacío
df = pd.DataFrame()
```

```python
producto = str(input("Ingrese el nombre del artículo: "))
cantidad = int(input("Ingrese la cantidad: "))
ventas = int(input("Precio unitario: "))
# Creación de variables y asignaciones
hoy=datetime.now()
hoy = hoy.strftime("%d %b %Y %H:%M:%S") #formato de la fecha día/mes/año/hora/minuto/segundos
#df['fecha'] = hoy
#df['producto'] = producto
#df['cantidad'] = cantidad
#df['ventas'] = ventas



# incorporando una nueva fila al DataFrame
nueva_fila = pd.DataFrame({
    'fecha': [hoy],
    'producto': [producto],
    'cantidad': [cantidad],
    'ventas': [ventas],
    'Antes': [cantidad * ventas],
    'Descontado': [cantidad * ventas * 0.95]
})

#Desindezando el DataFrame
df = df.reset_index()

# creamos un diccionario
df = pd.concat([df, nueva_fila], ignore_index = True)

#indexando la fecha
df.set_index('fecha', inplace = True)

#Eliminando columna index
if 'index' in df.columns:    
    df = df.drop(columns = 'index')

    
df
```

Para determinar un reporte con las ventas de los últimos 5 días, se considera que de la fecha actual, se resten **5** días y a partir de ahí generar una búsqueda de los elementos solicitados. 

```python
from datetime import timedelta
busqueda = datetime.today() - timedelta(5) #restando 5 días a la fecha actual
busqueda = busqueda.strftime("%d %b %Y %H:%M:%S") 

# La fecha con la cual se inicia la búsqueda es:
print("Se inicia la búsqueda desde: --", busqueda, " --al --", hoy)
```

```python
filtro = df.copy()
filtro = filtro.reset_index()
filtro = filtro[filtro['fecha'] >= busqueda]
filtro.set_index('fecha', inplace = True)
filtro
```

Finalmente, se generará un código para poder agrupar los productos generándose un resumen de cantidades y totales vendidos.


```python
resultado = df.groupby('producto')[['cantidad', 'Antes', 'Descontado']].sum()
resultado
```

## Ejercicio 3
Escribir programa que genere y muestre por pantalla un DataFrame con los datos de la siguiente tabla: 

```python
mes = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo']
Ventas = [30500, 35600, 28300, 33900, 42500]
Gastos = [22000, 23450, 18100, 35700, 32450]
tabla = pd.DataFrame(columns=['Mes', 'Ventas', 'Gastos'])
tabla['Mes'] = mes
tabla['Ventas'] = Ventas
tabla['Gastos'] = Gastos
tabla
```

A continuación, generar nuevas columna:

Columna 1 = Gasto/Ventas*100.  
Columna 2 = Si Ventas>= 30000, imprimir “Meta Superada”, caso contrario imprimir “Meta no Superada”.  
Columna 3 = Si Gasto>Ventas, calcular el sobre gasto.  
Obtener el total de las ventas de los últimos 3 meses  
Obtener el total de gastos de febrero a abril  

### Columna1
* Columna 1 = Gasto/Ventas*100.

```python
tabla['Columna1'] = tabla['Gastos'] / tabla['Ventas']*100
tabla
```

### Columna2
* Columna 2 = Si Ventas>= 30000, imprimir “Meta Superada”, caso contrario imprimir “Meta no Superada”.  


```python
#Generando la columna 2
for i in range (0, len(tabla)):
    if tabla.loc[i, 'Ventas'] >= 30000:
        tabla.loc[i,'Columna2'] = "Meta Superada"
    else:
        tabla.loc[i,'Columna2'] = "Meta no Superada"
tabla
```

### Columna3
* Columna 3 = Si Gasto - Ventas, calcular el sobre gasto [Ventas - Gastos]. Caso contrario `---`

```python
#Generando columna3
for i in range (0, len(tabla)):
    if tabla.loc [i, 'Gastos'] < tabla.loc [i, 'Ventas']:
        tabla.loc [i, 'Columna3'] =  tabla.loc [i, 'Ventas'] - tabla.loc [i, 'Gastos'] 
    else:
        tabla.loc [i, 'Columna3'] = "- "*5
tabla
```

### Obtener el total de las ventas de los últimos 3 meses


```python
inicio = len(tabla) - 3
fin = len(tabla)
tabla2 = tabla[inicio : fin]
tabla2
```

Y considerando las ventas totales de los últimos 3 meses:


```python
print("Las ventas de los últimos 3 meses es:" , tabla2['Ventas'].sum())
```

### Obtener el total de gastos de febrero a abril

```python
inicio = 1
fin = 4
tabla3 = tabla[inicio : fin]
tabla3
```

```python
print("Los gastos de los meses son:" , tabla3['Gastos'].sum())
```
